```python
#Question 1
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def isEmpty(self):
        return self.root is None

    def clear(self):
        self.root = None

    def insert(self, x):
        self.root = self._insert(self.root, x)

    def _insert(self, root, x):
        if root is None:
            return Node(x)
        else:
            if root.val < x:
                root.right = self._insert(root.right, x)
            elif root.val > x:
                root.left = self._insert(root.left, x)
        return root

    def search(self, x):
        return self._search(self.root, x)

    def _search(self, root, x):
        if root is None or root.val == x:
            return root
        if root.val < x:
            return self._search(root.right, x)
        return self._search(root.left, x)

    def breadth(self):
        if self.isEmpty():
            return []
        queue, traversal = [self.root], []
        while queue:
            node = queue.pop(0)
            traversal.append(node.val)
            if node.left: queue.append(node.left)
            if node.right: queue.append(node.right)
        return traversal

    def preorder(self, node):
        if node:
            print(node.val, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.val, end=' ')
            self.inorder(node.right)

    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.val, end=' ')

    def count(self):
        return self._count(self.root)

    def _count(self, node):
        if node is None:
            return 0
        else:
            return 1 + self._count(node.left) + self._count(node.right)

    def dele(self, x):
        self.root = self._dele(self.root, x)

    def _dele(self, root, x):
        if root is None:
            return root
        if x < root.val:
            root.left = self._dele(root.left, x)
        elif x > root.val:
            root.right = self._dele(root.right, x)
        else:
            if root.left is None:
                temp = root.right
                root = None
                return temp
            elif root.right is None:
                temp = root.left
                root = None
                return temp
            temp = self.minValueNode(root.right)
            root.val = temp.val
            root.right = self._dele(root.right, temp.val)
        return root

    def minValueNode(self, node):
        current = node
        while(current.left is not None):
            current = current.left
        return current

    def min(self):
        current = self.root
        while current.left is not None:
            current = current.left
        return current.val if current else None

    def max(self):
        current = self.root
        while current.right is not None:
            current = current.right
        return current.val if current else None

    def sum(self):
        return self._sum(self.root)

    def _sum(self, node):
        if node is None:
            return 0
        return node.val + self._sum(node.left) + self._sum(node.right)

    def avg(self):
        total_sum = self._sum(self.root)
        total_count = self._count(self.root)
        return total_sum / total_count if total_count > 0 else 0

    def height(self):
        return self._height(self.root)

    def _height(self, node):
        if node is None:
            return -1
        else:
            left_height = self._height(node.left)
            right_height = self._height(node.right)
            return 1 + max(left_height, right_height)

    def maxPathSum(self):
        return self._maxPathSum(self.root)

    def _maxPathSum(self, node):
        if node is None:
            return 0
        left = self._maxPathSum(node.left)
        right = self._maxPathSum(node.right)
        return node.val + max(left, right)

    def isAVL(self):
        return True 

    def mystery(self, node):
        if node is None:
            return 0
        else:
            return max(self.mystery(node.left), self.mystery(node.right))
    def isHeap(self):
        return True  
```

```python
#Question 2
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class StringBST:
    def __init__(self):
        self.root = None

    def isEmpty(self):
        return self.root is None

    def clear(self):
        self.root = None

    def search(self, value):
        return self._search(self.root, value)

    def _search(self, node, value):
        if node is None or node.value == value:
            return node
        elif value < node.value:
            return self._search(node.left, value)
        else:
            return self._search(node.right, value)

    def insert(self, value):
        if self.search(value) is None:
            self.root = self._insert(self.root, value)

    def _insert(self, node, value):
        if node is None:
            return TreeNode(value)
        elif value < node.value:
            node.left = self._insert(node.left, value)
        else:
            node.right = self._insert(node.right, value)
        return node

    def breadth(self):
        if self.isEmpty():
            return []
        queue, result = [self.root], []
        while queue:
            current = queue.pop(0)
            result.append(current.value)
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
        return result

    def preorder(self, node):
        if node:
            print(node.value, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.value, end=' ')
            self.inorder(node.right)

    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.value, end=' ')

    def count(self):
        return self._count(self.root)

    def _count(self, node):
        if node is None:
            return 0
        return 1 + self._count(node.left) + self._count(node.right)

    def dele(self, value):
        self.root = self._dele(self.root, value)

    def _dele(self, node, value):
        if node is None:
            return None
        if value < node.value:
            node.left = self._dele(node.left, value)
        elif value > node.value:
            node.right = self._dele(node.right, value)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            temp_val = self._findMin(node.right).value
            node.value = temp_val
            node.right = self._dele(node.right, temp_val)
        return node

    def _findMin(self, node):
        while node.left is not None:
            node = node.left
        return node

    def min(self):
        if self.isEmpty():
            return None
        else:
            return self._findMin(self.root).value

    def max(self):
        current = self.root
        while current.right is not None:
            current = current.right
        return current.value if current else None
    def sum(self):
        return None

    def avg(self):
        return None

    def height(self):
        return self._height(self.root)

    def _height(self, node):
        if node is None:
            return -1
        else:
            return 1 + max(self._height(node.left), self._height(node.right))

    def maxPathSum(self):
        return None

    def isAVL(self):
        return self._isAVL(self.root)[0]

    def _isAVL(self, node):
        if node is None:
            return True, -1
        left_balanced, left_height = self._isAVL(node.left)
        right_balanced, right_height = self._isAVL(node.right)
        node_height = 1 + max(left_height, right_height)
        balance = abs(left_height - right_height) <= 1
        return left_balanced and right_balanced and balance, node_height

    def mystery(self, node):
        if node is None:
            return 0
        else:
            return 1 + max(self.mystery(node.left), self.mystery(node.right))

    def isHeap(self):
        return None
```
